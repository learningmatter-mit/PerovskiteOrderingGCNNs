```python
import pandas as pd
```

```python
data_frame_names = ['edge_dataset','holdout_set_B_sites','holdout_set_series','test_set','training_set','validation_set']
data_es_names = ['edge_es','ternary_es','series_es','test_es','training_es','validation_es']

for i in range(len(data_frame_names)):
    data_frame = pd.read_json('data/' + data_frame_names[i] + '.json')
    data_es = pd.read_json('data/' + data_es_names[i] + '.json')
    assert data_frame.shape[0] == data_es.shape[0]
    data_frame_new = pd.concat([data_frame,data_es],axis=1)
    data_frame_new.to_json('data/' + data_frame_names[i] + '.json')
```
