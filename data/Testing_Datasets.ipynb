```python
import pandas as pd
import numpy as np
from pymatgen.core import Structure
```

```python
edge_data = pd.read_json("edge_dataset.json")

holdout_set_B_sites = pd.read_json("holdout_set_B_sites.json")

holdout_set_series = pd.read_json("holdout_set_series.json")

training_set = pd.read_json("training_set.json")

validation_set = pd.read_json("validation_set.json")

test_set = pd.read_json("test_set.json")
```

```python
##### Check that Unrelaxed Structures are Actually Unrelaxed

dataframes = [edge_data, holdout_set_B_sites,holdout_set_series,training_set,validation_set,test_set]

count = 0
for df in dataframes:
    for index, row in df.iterrows():
        unrelaxed_struct = Structure.from_dict(row.unrelaxed_struct)
        template_lattice = np.asarray([[8.0,0.0,0.0],[0.0,8.0,0.0],[0.0,0.0,8.0]])
        if ((unrelaxed_struct.lattice.matrix - template_lattice)>1e-6).any():
            print("Unrelaxed_Lattice_Error")
            
        coordinates = unrelaxed_struct.frac_coords
        
        for i in range(coordinates.shape[0]):
            for j in range(coordinates.shape[1]):
                min_dist = min(abs(coordinates[i,j]-0.0),
                               abs(coordinates[i,j]-0.25),
                               abs(coordinates[i,j]-0.5),
                               abs(coordinates[i,j]-0.75),
                                )
                if min_dist > 1e-5:
                    print(row.framework)
    count += 1
```

```python
### Check that all edges have 1 A and 1 B

for index, row in edge_data.iterrows():
    composition = row.composition
    if len(composition["sites"]["A"]) != 1 or len(composition["sites"]["B"]) != 1:
        print("Error in Composition")

### Check that all other structures have multiple

dataframes = [holdout_set_B_sites,holdout_set_series,training_set,validation_set,test_set]

for df in dataframes:
    for index, row in df.iterrows():
        composition = row.composition
        if len(composition["sites"]["A"]) < 2 and len(composition["sites"]["B"]) < 2:
            print("Error in Composition")
            
            
### Check that B_sites are 1 for A and 2 for B

dataframes = [holdout_set_B_sites]

for df in dataframes:
    for index, row in df.iterrows():
        composition = row.composition
        if len(composition["sites"]["A"]) != 1 or len(composition["sites"]["B"]) != 2:
            print("Error in Composition")
```

```python
### Check that formulas are disjoint sets

edge_formulas = set(edge_data.formula)
holdout_set_series_formulas = set(holdout_set_series.formula)
training_set_formulas = set(training_set.formula)
validation_set_formulas = set(validation_set.formula)
test_set_formulas = set(test_set.formula)
holdout_set_B_sites_formulas = set(holdout_set_B_sites.formula)


sets = [edge_formulas,holdout_set_series_formulas,training_set_formulas,
        validation_set_formulas,test_set_formulas,holdout_set_B_sites_formulas]

for i in range(len(sets)):
    for j in range(i+1,len(sets)):
        if not sets[i].isdisjoint(sets[j]):
            print(i)
            print(j)
            print("Overlapping Formulas")
```

```python
for formula in training_set_formulas:
    if formula in holdout_set_B_sites_formulas:
        print(formula)
```

```python
### Check for internal duplicate data - Formulas are disjoint ensures no leakage between datasets

from pymatgen.analysis.structure_matcher import StructureMatcher

SM = StructureMatcher()

dataframes = [holdout_set_B_sites,holdout_set_series,training_set,validation_set,test_set]

for df in dataframes:
    possible_formulas = set(df.formula)
    for formula in possible_formulas:
        
        ### grab similar formulas
        
        sub_df = df[df["formula"] == formula]
        
        for i in range(len(sub_df)):
            for j in range(i+1,len(sub_df)):
                struct_1 = Structure.from_dict(sub_df.iloc[i].unrelaxed_struct)
                struct_2 = Structure.from_dict(sub_df.iloc[j].unrelaxed_struct)
                if SM.get_rms_dist(struct_1,struct_2) is not None:
                    print("Issue with Duplication")
```
